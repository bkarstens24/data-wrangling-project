**Description:** This notebook is used for cleaning the raw data frames and .csv files from the "00" notebook.  The raw dataframes will be loaded in and then merged.  After the merge, any rows with duplicate track_ids are dropped as well as any rows that have a missing value for any column.  The drop duplicates for track_id is correctly ran twice to ensure the duplicates are dropped.

<br>

Import Libraries:

```python
import pandas as pd
```

```python
tophits = pd.read_csv('top1000songs_2010s_raw.csv', encoding='UTF-8')
apisongs = pd.read_csv('apitopsongs_2010s_raw.csv', encoding= 'UTF-8')
spotify = pd.read_csv('topsongstreams_raw.csv', encoding = 'UTF-8')

display(tophits)
display(apisongs)
display(spotify)
```

```python
merged_df = pd.merge(tophits, apisongs, on=['title', 'artist'], how='outer')
display(merged_df)
```

```python
final_merge_df = pd.merge(merged_df, spotify, on=['title', 'artist'], how='outer')
display(final_merge_df)
```

```python
final_song_df = final_merge_df.dropna()
display(final_song_df)
```

```python
final_song_df = final_song_df.drop_duplicates(subset=['track_id'], keep='first').drop_duplicates().reset_index(drop=True)
display(final_song_df)
```

```python
final_song_df = final_song_df.drop_duplicates(subset=['track_id'], keep='first').drop_duplicates().reset_index(drop=True)
display(final_song_df)
```

```python
print(final_song_df.dtypes)
```

```python
final_song_df['streams'] = final_song_df['streams'].replace({',': ''}, regex=True).astype(float)
```

```python
final_song_df.to_csv('final_song_clean.csv', encoding='UTF-8', index=False)
```
